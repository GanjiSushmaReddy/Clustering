# K-Means_Clustering

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing Datasets

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:,[3,4]].values
# We need two features to visualise the clusters. so we keep two columns(2-features)
## No train-test split, beacause no dependent variable
```

### Using the Elbow method to find the optimal no. of Clusters

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.xlabel('no. of clusters')
plt.ylabel('wcss')
plt.show()
# k=5
```

### Training the K-Means model on the dataset

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, init='k-means++',random_state=0)
y_kmeans = kmeans.fit_predict(x)
print(y_kmeans)
```

### Visualising the clusters

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1], s=80,color='r',label='cluster1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1], s=80,color='b',label='cluster2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1], s=80,color='g',label='cluster3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1], s=80,color='cyan',label='cluster4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1], s=80,color='magenta',label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,color='y',label='Centroids')
plt.title('K-Means clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending(1-100)')
plt.legend()
plt.show()
```
